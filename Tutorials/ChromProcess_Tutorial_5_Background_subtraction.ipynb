# ChromProcess Introduction, Part 4 (background subtraction)

This notebook illustrates how to perform a baseline substraction of a chromatogram by subtracting a 'blank' chromatogram from a sample chromatogram.

```python
import numpy as np
import matplotlib.pyplot as plt
from ChromProcess.Loading import chrom_from_csv
from ChromProcess.Loading import chrom_from_labsolutions_ascii
from ChromProcess.Processing import background_subtraction
```

# Load in the data.

```python
# NOTE: these two chromatograms were not recorded under the same conditions. 
#Â These data are for demonstration purposes only.
data_file = (
    "example_data/Example/ExperimentalData/ExampleChromatograms/chrom_001.csv"
)
background_chrom_file = (
    "example_data/Example/ExperimentalData/ExampleChromatograms/blank.csv"
)
sample_chrom = chrom_from_csv(data_file)
background_chrom = chrom_from_csv(background_chrom_file)
```

# Perform background subtraction

```python
background_corrected_chromatogram = background_subtraction(sample_chrom, background_chrom)
```

# Plot the results

```python
fig, ax = plt.subplots()
ax.plot(background_corrected_chromatogram.time, background_corrected_chromatogram.signal)
ax.plot(sample_chrom.time, sample_chrom.signal, c = "k")
ax.plot(background_chrom.time, background_chrom.signal, c = "r")
```
